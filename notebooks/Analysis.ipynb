## Specify the results to load

I'll start with plotting just one experiment. Later we may want to allow to plot multiple runs, like in Tensorboard.

```python
results_dir = '/global/cscratch1/sd/sfarrell/sc18-dl-tutorial/cifar10-cnn-N1-16150683'
```

```python
ls $results_dir
```

## Load the summary results and inspect them

```python
history = np.load(os.path.join(results_dir, 'history.npz'))
```

```python
n_ranks = int(history['n_ranks'])
```

```python
history.files
```

## Plot loss and accuracy

```python
fig, (ax0, ax1) = plt.subplots(nrows=1, ncols=2, figsize=(12, 5))

# Plot losses
ax0.plot(history['loss'], label='train')
ax0.plot(history['val_loss'], label='validation')
ax0.set_xlabel('Epoch')
ax0.set_ylabel('Loss')
ax0.legend(loc=0)

ax1.plot(history['acc'], label='train')
ax1.plot(history['val_acc'], label='validation')
ax1.set_xlabel('Epoch')
ax1.set_ylabel('Accuracy')
ax1.legend(loc=0)

plt.tight_layout()
```
